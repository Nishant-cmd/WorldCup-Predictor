```python
import pandas as pd

matches = pd.read_csv(
    "../../data/processed/training_data/feature_v1_dataset.csv"
)

importance_map = {

    # Friendlies
    "Friendly": 1,

    # Qualification tournaments
    "FIFA World Cup qualification": 3,
    "UEFA Euro qualification": 3,
    "African Cup of Nations qualification": 3,
    "AFC Asian Cup qualification": 3,
    "Gold Cup qualification": 3,
    "CONCACAF Nations League qualification": 3,

    #Nations League
    "UEFA Nations League": 3.5,
    "CONCACAF Nations League": 3.5,


    # Major continental tournaments
    "UEFA Euro": 4,
    "Copa América": 4,
    "African Cup of Nations": 4,
    "AFC Asian Cup": 4,
    "Gold Cup": 4,
    "UEFA Nations League": 4,
    "CONCACAF Nations League": 4,

    # Global tournaments
    "Confederations Cup": 5,
    "FIFA World Cup": 5
}
matches["tournament_importance"] = (
    matches["tournament"]
    .map(importance_map)
    .fillna(2)
)
print(
    matches["tournament_importance"]
    .value_counts()
)
matches["neutral"] = (
    matches["neutral"]
    .astype(int)
)
matches["home_advantage"] = (
    1 - matches["neutral"]
)
matches.to_csv(
    "../../data/processed/training_data/features_v2.csv",
    index=False
)
```
