```python
import pandas as pd

matches = pd.read_csv(
    "../../data/processed/training_data/features_v2.csv"
)

matches["date"] = pd.to_datetime(matches["date"])

matches = matches.sort_values("date").reset_index(drop=True)

h2h_history = {}
home_h2h_wins = []
away_h2h_wins = []
h2h_diff = []
for _, row in matches.iterrows():

    home = row["home_team"]
    away = row["away_team"]

    pair = tuple(sorted([home, away]))

    record = h2h_history.get(
        pair,
        {"draws": 0}
    )

    home_wins = record.get(home, 0)
    away_wins = record.get(away, 0)

    home_h2h_wins.append(home_wins)
    away_h2h_wins.append(away_wins)

    h2h_diff.append(
        home_wins - away_wins
    )

    home_goals = row["home_score"]
    away_goals = row["away_score"]

    if home_goals > away_goals:

        record[home] = (
            record.get(home, 0) + 1
        )

    elif away_goals > home_goals:

        record[away] = (
            record.get(away, 0) + 1
        )

    else:

        record["draws"] += 1

    
    h2h_history[pair] = record 
 
matches["home_h2h_wins"] = home_h2h_wins
matches["away_h2h_wins"] = away_h2h_wins

matches["h2h_diff"] = h2h_diff


matches.to_csv(
    "../../data/processed/training_data/features_v3.csv",
    index=False
)

def get_result(row):

    if row["home_score"] > row["away_score"]:
        return 2

    elif row["home_score"] < row["away_score"]:
        return 0

    else:
        return 1

matches["result"] = matches.apply(
    get_result,
    axis=1
)
print(matches["result"].value_counts())

print(
    matches["result"]
    .value_counts(normalize=True) * 100
)

matches.to_csv(
    "../../data/processed/training_data/final_training_dataset.csv",
    index=False
)

print(matches.columns.tolist())
```
