```python
import pandas as pd
import numpy as np

matches = pd.read_csv(
    "../../data/processed/match_data/matches_with_elo_fifa.csv"
)

matches["date"] = pd.to_datetime(matches["date"])

matches = matches.sort_values("date").reset_index(drop=True)
team_history = {}
home_form = []
away_form = []

for _, row in matches.iterrows():

    home = row["home_team"]
    away = row["away_team"]

    home_last5 = team_history.get(home, [])
    away_last5 = team_history.get(away, [])

    home_form.append(sum(home_last5[-5:]))
    away_form.append(sum(away_last5[-5:]))

    home_score = row["home_score"]
    away_score = row["away_score"]

    if home_score > away_score:

        home_points = 3
        away_points = 0

    elif home_score < away_score:

        home_points = 0
        away_points = 3

    else:

        home_points = 1
        away_points = 1

    team_history.setdefault(home, []).append(home_points)
    team_history.setdefault(away, []).append(away_points)


matches["home_form_5"] = home_form
matches["away_form_5"] = away_form

matches["form_diff"] = (
    matches["home_form_5"]
    - matches["away_form_5"]
)

print(
    matches[
        [
            "home_team",
            "away_team",
            "home_form_5",
            "away_form_5",
            "form_diff"
        ]
    ].head(20)
)
print(
    matches[
        [
            "home_form_5",
            "away_form_5",
            "form_diff"
        ]
    ].describe()
)
matches.to_csv(
    "../../data/processed/match_data/matches_with_form.csv",
    index=False
)
```
