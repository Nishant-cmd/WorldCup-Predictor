```python
import pandas as pd

historical_df=pd.read_csv('../../data/processed/match_data/historical_fifa_matches.csv')
fifa_df=pd.read_csv('../../data/processed/team_data/fifa_mens_rank_cleaned.csv')
elo_df=pd.read_csv('../../data/processed/team_data/elo_ratings_cleaned.csv')


historical_df["date"] = pd.to_datetime(
    historical_df["date"]
)
elo_df["date"] = pd.to_datetime(
    elo_df["date"]
)

fifa_df = fifa_df.rename(columns={
    "date": "year"
})

fifa_df["year"] = fifa_df["year"].astype(int)


match_teams = set(
    historical_df["home_team"]
).union(
    set(historical_df["away_team"])
)

elo_teams = set(elo_df["team"])


fifa_teams = set(fifa_df["team"])

missing_fifa = match_teams - fifa_teams

print(len(missing_fifa))
print(sorted(missing_fifa))

print("Unique dates:", elo_df["date"].nunique())

print((elo_df["rating"] == 0).sum())

print(elo_df["team"].value_counts().head(10))
elo_df[
    elo_df["team"] == "Argentina"
].sort_values("date")
```
