```python
import pandas as pd

# Load data
fifa_df = pd.read_csv("../../data/raw/fifa_mens_rank.csv")
worldcup_df=pd.read_csv('../../data/processed/worldcup_teams.csv')
historical_df=pd.read_csv('../../data/processed/historical_matches.csv')


# Remove unwanted columns
fifa_df = fifa_df.drop(columns=["semester", "acronym"])

# Standardize team names
fifa_df["team"] = fifa_df["team"].replace({
    "Korea Republic": "South Korea",
    "Côte d'Ivoire": "Ivory Coast",
    "USA": "United States",
    "Congo DR": "DR Congo",
    "IR Iran": "Iran",
    "Cape Verde Islands": "Cape Verde"
})

# Keep only years 2010–2024
fifa_df = fifa_df[(fifa_df["date"] >= 2010) & (fifa_df["date"] <= 2024)]

# Sort by year and rank
fifa_df = fifa_df.sort_values(by=["date", "rank"], ascending=[True, True])

# Reset index
fifa_df = fifa_df.reset_index(drop=True)
fifa_df.to_csv("../../data/processed/fifa_mens_rank_cleaned.csv", index=False)


print(fifa_df['team'].nunique())


worldcup_teams = set(worldcup_df["team"])

fifa_teams = set(fifa_df["team"])

missing_teams = worldcup_teams - fifa_teams


```

```python
historical_df=pd.read_csv('../../data/processed/historical_matches.csv')


valid_teams = set(fifa_df["team"])

historical_fifa_df = historical_df[
    historical_df["home_team"].isin(valid_teams)
    &
    historical_df["away_team"].isin(valid_teams)
]

historical_fifa_df.to_csv(
    "../../data/processed/historical_fifa_matches.csv",
    index=False
)

print("Original:", len(historical_df))
print("FIFA Only:", len(historical_fifa_df))

removed = len(historical_df) - len(historical_fifa_df)

print("Removed:", removed)


filtered_teams = set(
    historical_fifa_df["home_team"]
).union(
    set(historical_fifa_df["away_team"])
)

print(len(filtered_teams))
```
