```python
import pandas as pd

df = pd.read_csv("../data/raw/results.csv")

#Convert date column
df["date"]=pd.to_datetime(df["date"])

#Completed Matches
cutoff_date="2026-06-10"

historical_df = df.dropna(subset=["home_score", "away_score"])

historical_df = historical_df[
    historical_df["date"] <= cutoff_date
]

historical_df.to_csv("historical_matches.csv", index=False)

#Future Matches

future_df = df[
    (df["date"] > cutoff_date) &
    (df["home_score"].isna())
]

future_df.to_csv(
    "future_matches.csv",
    index=False
)


# Extract unique teams from future World Cup matches
worldcup_teams = sorted(
    set(future_df["home_team"]).union(
        set(future_df["away_team"])
    )
)

# Create dataframe
teams_df = pd.DataFrame({
    "team": worldcup_teams
})

# Save
teams_df.to_csv("worldcup_teams.csv", index=False)

print(f"Total World Cup Teams: {len(worldcup_teams)}")
print(teams_df.head())



```

```python
#Checking team name consistency
all_teams = sorted(
    set(historical_df["home_team"])
    .union(set(historical_df["away_team"]))
)

print(f"Total teams: {len(all_teams)}")
```

```python
pd.DataFrame({"team": all_teams}).to_csv(
    "all_teams.csv",
    index=False
)


historical_df["tournament"]\
    .value_counts()\
    .to_csv("tournament_distribution.csv")

historical_df["date"] = pd.to_datetime(
    historical_df["date"]
)

future_df["date"] = pd.to_datetime(
    future_df["date"]
)
print(historical_df["date"].min())
print(historical_df["date"].max())



```
